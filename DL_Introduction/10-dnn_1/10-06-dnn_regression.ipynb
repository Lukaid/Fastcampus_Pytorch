```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df["TARGET"] = boston.target
df.tail()
```

### StandardScaler
- feature의 범위를 표준화
- 비표준화된 feature는 딥러닝에 조금 더 치명적
- non-convex Loss surface는 딥러닝에서 치명적

```python
scaler = StandardScaler()
scaler.fit(df.values[:, :-1]) # 여기선 평균과 표준편차를 학습

df.values[:, :-1] = scaler.transform(df.values[:, :-1]) # 여기선 학습된 데이터로 표준화
# 근데 target은 표준화 안시키나?

df.tail()
```

## Train Model with PyTorch

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
data = torch.from_numpy(df.values).float()

data.shape
```

```python
y = data[:, -1:]
x = data[:, :-1]

print(x.shape, y.shape)
```

```python
n_epochs = 100000
learning_rate = 1e-4
print_interval = 5000
```

## Build Models

### Build Model using nn.Module

```python
relu = nn.ReLU()
leaky_relu = nn.LeakyReLU(0.1)
```

```python
class MyModel(nn.Module):
    
    def __init__(self, input_dim, output_dim): # 내가 필요한 것들을 미리 선언
        self.input_dim = input_dim
        self.output_dim = output_dim
        
        super().__init__()
        
        # 레이어를 만들어 둠
        self.linear1 = nn.Linear(input_dim, 3)
        self.linear2 = nn.Linear(3, 3)
        self.linear3 = nn.Linear(3, output_dim)
        self.act = nn.ReLU() # activation function 도 선언
        
    def forward(self, x): # 여기선 feed-forward를 구현
        # |x| = (batch_size, input_dim)
        h = self.act(self.linear1(x)) # |h| = (batch_size, 3)
        h = self.act(self.linear2(h))
        y = self.linear3(h) # 마지막 레이어에는 activation function 펑션 X
        # |y| = (batch_size, output_dim)
        
        return y
    
```

```python
model_1 = MyModel(x.size(-1), y.size(-1))

model_1
```

### Build Model with LeakyReLU using nn.Sequential

```python
optimizer_1 = optim.SGD(model_1.parameters(),
                      lr=learning_rate)
```

```python
for i in range(n_epochs):
    y_hat = model_1(x)
    loss_1 = F.mse_loss(y_hat, y)
    
    optimizer_1.zero_grad()
    loss_1.backward() # 얘가 미분
    
    optimizer_1.step() # 파라미터 업데이트
    
    if (i + 1) % print_interval == 0:
        print('Epoch %d: loss=%.4e' % (i + 1, loss_1))
```

### Result

```python
df_1 = pd.DataFrame(torch.cat([y, y_hat], dim=1).detach().numpy(),
                  columns=["y", "y_hat"])

sns.pairplot(df_1, height=5)
plt.show()
```

## 더 간단히 모델 선언하는 방법

```python
model_2 = nn.Sequential(
    nn.Linear(x.size(-1), 3),
    nn.LeakyReLU(),
    nn.Linear(3, 3),
    nn.LeakyReLU(),
    nn.Linear(3, 3),
    nn.LeakyReLU(),
    nn.Linear(3, 3),
    nn.LeakyReLU(),
    nn.Linear(3, 3),
    nn.LeakyReLU(),
    nn.Linear(3, y.size(-1)),
)

model_2
```

Grediant descent, 대상은 모델의 파라미터

```python
optimizer_2 = optim.SGD(model_2.parameters(),
                      lr=learning_rate)
```

```python
for i in range(n_epochs):
    y_hat = model_2(x)
    loss_2 = F.mse_loss(y_hat, y)
    
    optimizer_2.zero_grad()
    loss_2.backward() # 얘가 미분
    
    optimizer_2.step() # 파라미터 업데이트
    
    if (i + 1) % print_interval == 0:
        print('Epoch %d: loss=%.4e' % (i + 1, loss_2))
```

## Let's see the result!

```python
df_2 = pd.DataFrame(torch.cat([y, y_hat], dim=1).detach().numpy(),
                  columns=["y", "y_hat"])

sns.pairplot(df_2, height=5)
plt.show()
```
